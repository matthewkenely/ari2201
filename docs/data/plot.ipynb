```python
df = pd.read_csv('./location_articles_images.csv')
```

```python
all_locs = []

for l in df['location']:
    for i in eval(l):
        all_locs.append(i)
```

```python
loc_freqs = {}

for i in all_locs:
    if i in loc_freqs:
        loc_freqs[i] += 1
    else:
        loc_freqs[i] = 1
```

```python
loc_freqs = {k: v for k, v in sorted(loc_freqs.items(), key=lambda item: item[1], reverse=True)}

# bar plot plt
plt.bar(list(loc_freqs.keys())[:59], list(loc_freqs.values())[:59])

# rotate x axis labels
plt.xticks(rotation=90)

# increase fig size
plt.rcParams["figure.figsize"] = (12, 6)

plt.show()
```
